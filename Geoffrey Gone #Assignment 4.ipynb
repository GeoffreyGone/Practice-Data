```python
#importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

pd.options.display.max_columns=50
```

```python
#Reading data into DataFrame
carsales_df = pd.read_csv("C:/Users/Dela/Desktop/Blossom Academy/carsales.csv", index_col=0)
carsales_df.head(10)
```

## Data Cleaning
Looking through the data, it has no missing values but some rows contains certain inappropriate characters that signifies that those row values were not recorded well.

```python
# Replace all rows containing '?' to NaN
carsales_df = carsales_df.replace("?", np.nan)
carsales_df.head()
```

```python
# Exploring the data columns to identify non-null values
carsales_df.info()
```

```python
carsales_df.isnull().sum()
```

```python
#Dealing with missing values in the 'price' column
price_clean = carsales_df[carsales_df['price'].isnull()]
missing_price = len(price_clean)
total = len(carsales_df)
percent_price_mssing = missing_price / total * 100
print(" {:.2f} of the data is missing in the 'price' column".format(percent_price_mssing))
```

```python
# Droping missing rows in the 'price' column
carsales_df = carsales_df.dropna(how='any', subset=['price'])
carsales_df.shape
```

```python
#Using the fillna method 'bfill' to replace missing values in 'num-of-doors'
carsales_df['num-of-doors'] = carsales_df['num-of-doors'].fillna(method= 'bfill')
```

```python
#Exploring the 'normalized-losses' as it contains the most missing values
normalized = carsales_df[carsales_df['normalized-losses'].isnull()]
normalized
```

```python
missing_nor = len(normalized)
total = len(carsales_df)
percent_mssing = missing_nor / total * 100
print("{}% of the data is missing in the 'normalized-losses' column".format(percent_mssing))
```

Looking at the data, a large number representing 18.4% is missing. Since this column is numeric and a predictor value, the missing rows in the normalized column will be filled with the mean value. 

#### Converting the numerical variables which were stored as objects into numerical variables and filling with the mean

```python
#Numeric values that were stored as objects
numeric = ['normalized-losses','bore', 'stroke', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
```

```python
carsales_df[numeric] = carsales_df[numeric].astype(float)
```

```python
#Fill the missing columns with the column mean
num = ['normalized-losses','bore', 'stroke', 'horsepower', 'peak-rpm']
carsales_df[num] = carsales_df[num].fillna(carsales_df[num].mean())
```

```python
carsales_df.shape
```

```python
carsales_df.head()
```

```python
carsales_df.corr()
```

```python
correlation = carsales_df.corr()
top_corr_cols = correlation.price.sort_values(ascending=False).keys()
top_corr = correlation.loc[top_corr_cols, top_corr_cols]
dropSelf = np.zeros_like(top_corr)
dropSelf[np.triu_indices_from(dropSelf)] = True
plt.figure(figsize=(18, 10))
sns.heatmap(correlation, cmap=sns.diverging_palette(220, 10, as_cmap=True), annot=True, fmt=".2f", mask=dropSelf)
sns.set(font_scale=1.5)
plt.show()
del correlation, dropSelf, top_corr
```

From the correlation matrix above, car properties such as symboling, peak-rpm, city-mpg and highway-mpg are negatively correlated with the car price, This means that as price increases, the value of  symboling, peak-rpm, city-mpg and highway-mpg decreases.

Car properties such as normalized-losses, wheel-base, length, width, height, curb-weight, engine-size, bore, stroke, compression-ratio and horsepower are positively correlated with price. This means that as price increases, these components also increases.

The first five strongly correlated variables with price are engine-size, horsepower, curb-weight, width, length.

```python
sns.pairplot(carsales_df, x_vars=['engine-size', 'highway-mpg', 'peak-rpm'], 
             y_vars='price', 
             size=7, 
             kind='reg')
```

```python
#boxplot
sns.set(rc={"axes.grid": False,
            "figure.figsize": (10.0, 7.0),
            "axes.labelsize": 15,
            "xtick.labelsize": 15,
            "ytick.labelsize": 15
           })

sns.boxplot(x='price', y='body-style', data=carsales_df, 
            linewidth=1.5, 
            width=0.5,
            fliersize=6.0)
```

The boxplot above shows that cars whose body-style is hardtop have their minimum price around 10,000 and the maximum around 45,000. Its median price is also around 20,000.

Sadan body-style cars have majority of their prices within the third quartile and have some price values as outliers.

```python
#drive-wheels type that has the highest average price
#Using the groupby method 
carsales_df.groupby('drive-wheels')['price'].mean().sort_values(ascending=False)
```

On average, cars with rwd (rear-wheel drive) are the more expensive as compared to 4wd (four-wheel drive) and fwd (front-wheel drive)

```python
# Creating 3 bins to categorize the horsepower of cars into cars with Low, Medium and High Horsepower.
carsales_df['binned_hp'] = pd.cut(carsales_df["horsepower"], 3, include_lowest=True, labels=['Low', 'Medium', 'High'])
carsales_df.head()
```

```python
#Using the binned horsepower to draw a bar chat
fig, ax = plt.subplots(figsize=(10, 6))
ax = carsales_df['binned_hp'].value_counts().plot(kind='bar')
ax.set_xlabel("Horsepower category")
ax.set_ylabel("Frequency")
ax.set_title("Car Horsepower")
plt.show()
```

From the graph above, majority (153) of the cars included in the car sales dataset have low horsepower with few of the cars within the medium range (43) and the high range (5)
