```python
import pandas as pd
import numpy as np
```

```python
test_df = pd.read_csv("test.csv")
train_df = pd.read_csv("train.csv")
```

```python
train_df.head(10)
```

```python
train_df.drop("Loan_ID", axis=1, inplace=True)
```

```python
test_df.head(10)
```

```python
test_df.drop("Loan_ID", axis=1, inplace=True)
```

```python
train_df.isnull().sum()
```

```python
train_data = train_df[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History', 'Loan_Status']]
train_data.head()
```

```python
train_data.isnull().sum()
```

```python
missing_ = ['LoanAmount', 'Loan_Amount_Term', 'Credit_History']
train_data[missing_] = train_data[missing_].fillna(train_data[missing_].mean())
```

```python
train_data.isnull().sum()
```

```python
X = train_data.iloc[:, :-1].values
Y = train_data.iloc[:, -1].values
```

```python
X
```

```python
Y
```

```python
#Encoding dependent variable
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
Y = le.fit_transform(Y)
```

```python
Y
```

```python
#splitting dataset into train and test set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=0)
```

```python
X_train
```

```python
X_test
```

```python
Y_train 
```

```python
Y_test
```

```python
# Fitting Logistic Regression to Training set
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0, solver='lbfgs')
classifier.fit(X_train, Y_train)
```

```python
# Predicting Test set results
Y_pred = classifier.predict(X_test)
```
